```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset
file_path = 'air_pollution_lung_cancer_new.csv'
data = pd.read_csv(file_path)

# Display the first few rows and basic information about the dataset
data.head(), data.info(), data.describe()
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
```

```python
# Visualization before trimming
data.boxplot(column=['Lung Cancer', 'PM2.5', 'PM10'])
plt.title('Box Plot of Original Data')
plt.show()

data[['Lung Cancer', 'PM2.5', 'PM10']].hist(bins=30, figsize=(10, 5))
plt.suptitle('Histogram of Original Data')
plt.show()
```

```python
Q1 = data.quantile(0.25, numeric_only=True)
Q3 = data.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
aligned_data = data.select_dtypes(include=['float64', 'int64'])
outliers = ((aligned_data < (Q1 - 1.5 * IQR)) | (aligned_data > (Q3 + 1.5 * IQR)))
print("\nOutliers Detected:")
print(outliers.sum())
```

```python
# Trimming outliers
t_data = data[~((aligned_data < (Q1 - 1.5 * IQR)) | (aligned_data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Summary statistics after trimming
trimmed_summary = t_data.describe()
print("Trimmed Summary Statistics:")
print(trimmed_summary)
```

```python
# Visualization after trimming
t_data.boxplot(column=['Lung Cancer', 'PM2.5', 'PM10'])
plt.title('Box Plot of Trimmed Data')
plt.show()

t_data[['Lung Cancer', 'PM2.5', 'PM10']].hist(bins=30, figsize=(10, 5))
plt.suptitle('Histogram of Trimmed Data')
plt.show()
```
